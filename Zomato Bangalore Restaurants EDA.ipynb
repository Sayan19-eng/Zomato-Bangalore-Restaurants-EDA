## Exploratory Data Analysis : Zomato Bangalore Restaurants

-  The data is accurate to that available on the zomato website until 15 March 2019.

####  #Importing Dataset

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Zomato_reduced.csv')
```

```python
df
```

####  #Data Preparation and Cleaning

- Looked at some information about data & columns.
- Fixed any missing or incorrect values.
- Dropped columns that have no or very less impact on our analysis.
- Renamed some columns.
- Splitted the column data where required.

```python
df.info()
```

```python
#statistical analysis
df.describe()
```

```python
#checking null values
df.isnull().sum()
```

```python
# removing columns that have less or no impact on the analysis
df = df.drop(['url','phone','address','dish_liked','reviews_list','menus_combined','location_latitude',
              'location_longitude'],axis=1)
```

```python
# renaming some columns
df = df.rename(columns={'listed_in(type)':'meal_type'})
df = df.rename(columns={'listed_in(city)':'area'})
df = df.rename(columns={'approx_cost(for two people)':'approx_cost_for_2'})
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#splitting the entries in cuisines and rest_type
df['cuisines']= df['cuisines'].str.split(",", n = 10, expand = True)
df['rest_type']= df['rest_type'].str.split(",", n = 10, expand = True)
```

```python
df.head()
```

  #### #Exploratory Analysis and Visualization

####  Area having highest and lowest number of restaurants

```python
df.area.nunique()
```

```python
df.area.unique()
```

```python
restaurants_by_area = df.area.value_counts()
```

```python
restaurants_by_area
```

```python
restaurants_by_area.plot(kind='bar',color=['#D9514EFF','#435E55FF'],figsize=(15,8))
plt.title("Number of restaurants in each area")
plt.ylabel("Number of restaurants")
```

#### Number of customers in each area(who have given ratings)

```python
def cm_to_inch(value):
    return value/2.54

plt.figure(figsize=(cm_to_inch(30), cm_to_inch(20)))
plt.barh(df.area, df.votes, color = ["#D9514EFF", "#435E55FF"])
plt.title('Customer Rating vs Area')
plt.xlabel('Number of customers who have given rating')
plt.ylabel('Area')
```

#### Table Booking Option

```python
book_or_not = df.book_table.value_counts()
```

```python
book_or_not
```

```python
labels = ['NO', 'YES']
myexplode = [0, 0.2]
mycolors = ["#D9514EFF", "#435E55FF"]
plt.pie(book_or_not, labels=labels, startangle=60, explode = myexplode, shadow=True, colors = mycolors,autopct='%1.2f%%')
plt.title("How many restaurants allow table booking?")
fig = plt.gcf()
fig.set_size_inches(6, 6)
plt.legend() 
```

#### Restaurants accept online orders or not?

```python
online_order_or_not = df.online_order.value_counts()
online_order_or_not
```

```python
labels = ['YES', 'NO']
myexplode = [0.1, 0]
mycolors = ["#435E55FF", "#D9514EFF"]
plt.pie(online_order_or_not, labels=labels, startangle=90, explode = myexplode, shadow=True, colors = mycolors,autopct='%1.2f%%')
plt.title("How many restaurants accept online orders?")
fig = plt.gcf()
fig.set_size_inches(6, 6)
plt.legend() 
```

#### # Summary and Conclusion

- 'BTM' have highest number of restaurants followed by 'Brigade Road' and 'Bannerghatta Road'. Meanwhile 'Residency Road' have least number of restaurants.
- 'Bellandur' have highest number of customers followed by 'Indiranagar' and 'BTM'.
- 'North Indian' cuisine is widely available (2267 restaurants) in Bangalore followed by 'South indian' (972 restaurants) and 'Biryani' (563 restaurants).
- 9.94% of restaurants i.e 920 out of 9253 restaurants allow table booking in Bangalore.
- 61.41% of restaurants i.e 5682 out of 9253 restaurants accept online orders in Bangalore.
- 'BTM' have 9 high rated restaurants (rating > 4.5) followed by 'Indiranagar' and 'Brookefield' with 8 and 7 restaurants respectively.
- 'Byg Brewski Brewing Company', 'Toit', 'Truffles', 'AB's - Absolute Barbecues', 'The Black Pearl' are the best restaurants with rating > 4.6 and votes > 10,000.
- 'CTR', 'Brahmin's Coffee Bar', 'O.G. Variar & Sons', 'Taaza Thindi' are the highest rated restaurants with average cost for two person less than or equal to 200 Rs.
- 'Byg Brewski Brewing Company' is the best restaurant in the city with highest rating by highest number of customers with average cost for two person = 1600 Rs.
- 'Truffles' in 'Koramangala 5th Block' is the best cafe in city with average rating of 4.7 by 14654 people.
- 'Punjab Grill' and 'The Black Pearl' are the best North Indian restaurant franchise in the city.
- 'CTR' is the best South Indian restaurant in the city.
- 'ECHOES Koramangala' is the best Chinese restaurant in the city.
- Most of the restaurants in Bangalore are 'Quick Bites' (3504 restaurants) followed by 'Casual Dining' (2235 restaurants) then 'Cafe' (552 restaurants).
- Out of 9253 restaurants in Bangalore, 528 restaurants do delivery.
